```python
from src.LLMs.load_llm import gpt_oss_20b
import requests
from langchain_core.messages import HumanMessage,SystemMessage
```

```python
url = "http://127.0.0.1:8000/tools/expensetracker"
response = requests.get(url)
tools = response.json()
```

```python
llm = gpt_oss_20b().bind_tools(tools=tools)
```

```python
response = llm.invoke([
    SystemMessage(content="user tools if needed for any task else response normaly. and if any tool or work requires date then use get_CurrentDate tool which will give you current date then use it anywhere."),
    HumanMessage(content="today i bought  bananas for 50rs ")
],config={"congfigurable":{"user_id":"2121"}})
```
